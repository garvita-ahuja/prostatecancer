```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load the dataset into a Pandas DataFrame
data = pd.read_csv(r'C:\Users\GARVITA\Downloads\PCa_data.csv')
data
```

```python
data.dropna(how='all', inplace=True)
# Replace missing values with the median of the column
data['Family'] = data['Family'].fillna("No")
data['BP'] = data['BP'].fillna("No")
data['Class'] = data['Class'].fillna("No")
data['Bones'] = data['Bones'].fillna("No")
data['Nodes'] = data['Nodes'].fillna("No")
data['Viscera'] = data['Viscera'].fillna("No")
data['Gleason']=data['Gleason'].fillna(0)
data['Nadir PSA']=data['Nadir PSA'].fillna(0)
data['ECOG']=data['ECOG'].fillna(data['ECOG'].mean())
data['PSA']=data['PSA'].fillna(data['ECOG'].mean())
data['Second'] = data['Second'].fillna("No")
data['Third'] = data['Third'].fillna("No")
data['Fourth'] = data['Fourth'].fillna("No")

dummies = pd.get_dummies(data[['Family', 'Histology', 'Bones', 'Nodes', 'Viscera','M', 'N', 'BP', 
                               'Class', 'First', 'Second', 'Third', 'Fourth']])
data = pd.concat([data, dummies], axis=1)
data.columns
```

```python
import pandas as pd

seq_count = {}

for i, row in data.iterrows():
    sequence = row[['First', 'Second', 'Third', 'Fourth']].dropna().tolist()
    sequence = ','.join(sequence)
  
    if sequence not in seq_count:
        seq_count[sequence] = {'count': 0, 'survived': 0}
    seq_count[sequence]['count'] += 1
    
    if row['Class'] == 'yes':
        seq_count[sequence]['survived'] += 1

seq_class_df = pd.DataFrame.from_dict(seq_count, orient='index')
seq_class_df = seq_class_df.sort_values(by='count', ascending=False)

seq_class_df['survival_rate'] = (seq_class_df['survived'] / seq_class_df['count']) * 100

print(seq_class_df[['survived', 'count', 'survival_rate']].head(10))
```
